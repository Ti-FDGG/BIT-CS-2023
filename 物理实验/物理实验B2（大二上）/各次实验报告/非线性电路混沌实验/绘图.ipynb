```python
import pandas as pd
from matplotlib import pyplot as plt

# 显示中文和负号
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

root_address = r"C:\Users\Timothy\Desktop\学习资料\物理实验\物理实验B2（大二上）\各次实验报告\非线性电路混沌实验"
address = root_address + r"\数据处理.xlsx"

df = pd.read_excel(address, sheet_name="Sheet1")
```

```python
# 提取U(V)和-I(mA)
x = df["U(V)"]
y = df["-I(mA)"]

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', markersize=4, linestyle='-')

# 设置图例和标签
plt.xlabel("U(V)", fontsize=14)
plt.ylabel("-I(mA)", fontsize=14)
plt.title("有源非线性电阻的伏安特性曲线", fontsize=20)
plt.grid(True)
plt.savefig(root_address + r"\有源非线性电阻的伏安特性曲线.png")
```
