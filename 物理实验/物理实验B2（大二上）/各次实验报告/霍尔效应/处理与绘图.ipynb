```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 定义数据
IM = np.array([0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1])
B = np.array([340, 293, 258, 213, 168, 125, 80, 39])

# 绘制散点图
plt.scatter(IM, B, color='blue', label='数据点')

# 进行线性拟合
coefficients = np.polyfit(IM, B, 1)
linear_fit = np.poly1d(coefficients)

# 绘制拟合的直线
plt.plot(IM, linear_fit(IM), color='red', label='拟合直线')

# 添加标签和标题
# plt.xlabel('IM (A)')
# plt.ylabel('B (mT)')
# plt.title('IM-B 散点图及拟合直线')
# plt.legend()

# 显示图形
plt.show()
```
