```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import linregress

# 显示中文和负号
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

root_address = r"C:\Users\Timothy\Desktop\学习资料\物理实验\物理实验B2（大二上）\各次实验报告\光电效应普朗克常数"
address = root_address + r"\数据处理.xlsx"

df1 = pd.read_excel(address, sheet_name="1")
df2_1 = pd.read_excel(address, sheet_name="2_1")
df2_2 = pd.read_excel(address, sheet_name="2_2")
df3_2 = pd.read_excel(address, sheet_name="3_2")
```

```python
# 提取频率和Uc拔
x = df1["频率ν(×10^14Hz)"]
y = df1["Uc拔(V)"]

# 进行线性回归拟合
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# 计算拟合直线的y值
y_fit = slope * x + intercept

# 绘制数据点和拟合直线
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="数据点", marker='^', s=80)
plt.plot(x, y_fit, color='red', label=f"拟合直线: y = {slope:.3f}x + {intercept:.3f}")

# 设置图例和标签
plt.xlabel("ν(×10^14Hz)", fontsize=14)
plt.ylabel("Uc(V)", fontsize=14)
plt.title("Uc-ν关系曲线", fontsize=20)
plt.grid(True)
plt.savefig(root_address + r"\图\1-Uc-ν关系曲线.png")

# 输出斜率和截距
print(f"斜率: {slope}")
print(f"截距: {intercept}")
```

```python
# 提取UAC(V)和I(×10^-11A)
x = df2_1["UAC(V)"]
y = df2_1["I(×10^-11A)"]

# 绘制伏安特性曲线
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='^', linestyle='-', label="伏安特性曲线")

# 设置图例和标签
plt.xlabel("UAC(V)", fontsize=14)
plt.ylabel("I(×10^-11A)", fontsize=14)
plt.title("伏安特性曲线", fontsize=20)
plt.legend()
plt.grid(True)
plt.savefig(root_address + r"\图\2-伏安特性曲线.png")
```
